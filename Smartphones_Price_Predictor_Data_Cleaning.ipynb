### 1. Model :

```python
# extracting brand name from the model
df['brand'] = [x[0] for x in df['model'].str.split()]
```

```python
df.sample(5)
```

```python
print(len(df.brand.unique()))
df.brand.unique()
```

```python
# 'Redmi' and 'Xiaomi' sort of refer to same brand
df[df['brand']=='Redmi']
```

```python
# function for replacing more than one elements
def series_replace(series,old,new):
    n=len(old)
    for i in range(n):
        series = series.str.replace(old[i],new[i])
    
    return series
```

```python
# fixing 'Poco', 'Oppo' and 'Redmi'
old = ['Poco','Oppo','Redmi']
new = ['POCO','OPPO','Xiaomi']
df['brand'] = series_replace(df['brand'],old,new)
```

### 4. Processor :

```python
df.head(2)
```

##### A. Processor_name

```python
# processor name 
temp_3 = df.processor.str.split(',',expand=True)
```

```python
temp_3.head()
```

```python
# replacing similar processors
old=['A13','Apple','SC9863A','Spreadtrum','Sanpdragon','SC6531E','Samsung','Qualcomm']
new=['Bionic','Bionic','Unisoc','Unisoc','Snapdragon','Unisoc','Exynos','Snapdragon']
df['processor_name'] = series_replace(temp_3[0].str.split(expand=True)[0],old,new)
```

```python
df.head(2)
```

```python
# function for grouping some items of a pandas series
def series_grouping(word,items,group_name):
    if word in items:
        return group_name
    else:
        return word
```

```python
# replacing unneccessary items with 'None'
items=['Octa','No','Dual','Quad','1','4','1450','32','800','2000','52','1.77','(28','Single','256','48','8']
df['processor_name']=df['processor_name'].apply(series_grouping,items=items,group_name=np.nan)
```

##### B. Cores in Processor

```python
# function for calculating number of cores in the processor
def no_of_cores(word):
    if 'Octa' in word:
        return 8
    elif 'Hexa' in word:
        return 6
    elif 'Quad' in word:
        return 4
    elif 'Dual' in word:
        return 2
    else:
        return 'None'
```

```python
df['processor_cores'] = temp_3[1].fillna('None').apply(no_of_cores).replace('None',np.nan)
```

```python
print(temp_3[2].unique())
temp_3[2].isnull().sum()
```

```python
# processor speed column 
df['processor_speed'] = temp_3[2].str.replace('GHz Processor','')
```

### 5. Ram and Storage:

```python
df.head(2)
```

```python
# missing values
df.ram.isnull().sum()
```

```python
df['storage'] = df.ram.str.split(',',expand=True)[1]
df['ram'] = df.ram.str.split(',',expand=True)[0]
```

```python
df.head(2)
```

```python
# null values in 'ram' and 'storage' columns
print(df['ram'].isnull().sum())
print(df['storage'].isnull().sum())
```

```python
# 'ram' column and 'storage' column
df['ram'] = df['ram'].apply(lambda x:x if 'RAM' in x else np.nan)
df['storage'] = df['storage'].fillna('').apply(lambda x:x if 'inbuilt' in x else np.nan)
```

### 6. Battery :

```python
df['battery_capacity'] = df['battery'].str.split('Battery',expand=True)[0].apply(lambda x:x if 'mAh' in x else np.nan)
df['battery_power'] = df['battery'].str.split('Battery',expand=True)[1].str.split(expand=True)[1].fillna('None').apply(lambda x:x if 'W' in x else np.nan)
```

```python
df.head(2)
```

```python
# null values in 'ram' and 'storage' columns
print(df['ram'].isnull().sum())
print(df['storage'].isnull().sum())
```

### 7. Display :

```python
df.display.str.replace('\u2009',' ').unique()
```

```python
# temporary variable for display_size and display_resolution
temp_4 = df['display'].str.split('px',expand=True)[0].str.split(',',expand=True)
```

```python
df['display_size'] = temp_4[0].apply(lambda x:x if 'inches' in x else np.nan).str.replace(' inches','').astype(np.float64)
df['display_resolution'] = temp_4[1].fillna('').apply(lambda x:x if 'x' in x else np.nan).str.replace('\u2009','')
```

```python
df.head(3)
```

### 8. Camera :

```python
df['camera'] = df['camera'].str.replace('\u2009',' ')
```

```python
# temporary variable for camera 
temp_5 = df['camera'].str.split('&',expand=True)
```

```python
df['rear_camera'] = temp_5[0].str.split('+',expand=True)[0].fillna('').apply(lambda x:x.split('MP')[0] if 'MP' in x else np.nan)
df['front_camera'] = temp_5[1].fillna('').apply(lambda x:x.split('MP')[0] if 'MP' in x else np.nan)
```

```python
df.columns
```

### 9. Card :

```python
print(df['card'].isnull().sum())
df['card'].value_counts().index
```

```python
# variable for card column
card_series = df['card'].fillna('').apply(lambda x:x if 'Memory Card' in x else 'None')
```

```python
memory_card_supported = []
for word in card_series:
    if 'Not' in word:
        memory_card_supported.append(0)
    elif 'None' in word:
        memory_card_supported.append(np.nan)
    else:
        memory_card_supported.append(1)
        
df['memory_card_supported']=memory_card_supported
```

```python
df.head(3)
```

```python
df['os'].value_counts()
```

```python
# operating system
df['os'] = df['os'].fillna('').apply(lambda x:x if 'v' in x else np.nan)
```
